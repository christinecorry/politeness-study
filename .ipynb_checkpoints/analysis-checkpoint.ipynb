# Politeness Study — Analysis
Run `collect.py` first to generate `data/results.json`.

```python
import json
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np

with open('data/results.json') as f:
    data = json.load(f)

df = pd.json_normalize(data, sep='_')
print(df.shape)
df.head()
```

```python
# Average scores by condition
dims = ['scores_accuracy', 'scores_completeness', 'scores_clarity', 'scores_depth']
df[dims] = df[dims].apply(pd.to_numeric, errors='coerce')
df.groupby('condition')[dims].mean().round(2)
```

```python
for dim in dims:
    print(f"\n{dim}:")
    print(df.groupby('condition')[dim].value_counts().sort_index())
```

```python
# Paired t-test per dimension
# Aggregate runs by taking the mean score per prompt/condition first
agg = df.groupby(['prompt_id', 'condition'])[dims].mean().reset_index()
polite = agg[agg['condition'] == 'polite'].set_index('prompt_id')
blunt  = agg[agg['condition'] == 'blunt'].set_index('prompt_id')

print('Paired t-tests (polite vs blunt):\n')
for dim in dims:
    t, p = stats.ttest_rel(polite[dim], blunt[dim])
    d = (polite[dim] - blunt[dim]).mean() / (polite[dim] - blunt[dim]).std()
    print(f'{dim:30s}  t={t:.3f}  p={p:.3f}  Cohen\'s d={d:.3f}')
```

```python
# Bar chart: mean scores by condition and dimension
means = df.groupby('condition')[dims].mean()
means.index = ['Blunt', 'Polite']
means.columns = ['Accuracy', 'Completeness', 'Clarity', 'Depth']

means.T.plot(kind='bar', figsize=(8, 5), color=['#e07b54', '#5b8db8'])
plt.title('Mean Judge Scores by Condition')
plt.ylabel('Score (1–5)')
plt.ylim(1, 5)
plt.xticks(rotation=0)
plt.legend(title='Condition')
plt.tight_layout()
plt.savefig('data/scores_by_condition.png', dpi=150)
plt.show()
```

```python
# Break down by domain
df.groupby(['domain', 'condition'])[dims].mean().round(2)
```

```python
# ── Cell 1: Overall bar chart (excluding clarity) ────────────────────────────
 
valid_dims = ['scores_accuracy', 'scores_completeness', 'scores_depth']
dim_labels = ['Accuracy', 'Completeness', 'Depth']
 
means = df.groupby('condition')[valid_dims].mean()
means.index = ['Blunt', 'Polite']
means.columns = dim_labels
 
fig, ax = plt.subplots(figsize=(8, 5))
x = np.arange(len(dim_labels))
width = 0.35
 
bars1 = ax.bar(x - width/2, means.loc['Blunt'],  width, label='Blunt',  color='#e07b54', alpha=0.85)
bars2 = ax.bar(x + width/2, means.loc['Polite'], width, label='Polite', color='#5b8db8', alpha=0.85)
 
ax.set_title('Mean Judge Scores by Condition (excluding clarity — ceiling effect)', fontsize=12)
ax.set_ylabel('Mean Score (1–5)')
ax.set_xticks(x)
ax.set_xticklabels(dim_labels)
ax.set_ylim(3.5, 5.2)
ax.legend(title='Condition')
ax.axhline(y=5, color='gray', linestyle='--', alpha=0.4, label='Max score')
 
# Add value labels on bars
for bar in bars1:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02,
            f'{bar.get_height():.2f}', ha='center', va='bottom', fontsize=9)
for bar in bars2:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02,
            f'{bar.get_height():.2f}', ha='center', va='bottom', fontsize=9)
 
plt.tight_layout()
plt.savefig('data/overall_scores.png', dpi=150)
plt.show()
 
```

```python
# ── Cell 2: Split by domain ───────────────────────────────────────────────────
 
domains = df['domain'].unique()
fig, axes = plt.subplots(1, len(domains), figsize=(14, 5), sharey=True)
fig.suptitle('Mean Judge Scores by Domain and Condition', fontsize=13)
 
for ax, domain in zip(axes, sorted(domains)):
    subset = df[df['domain'] == domain].groupby('condition')[valid_dims].mean()
    # Ensure both conditions are present
    for cond in ['blunt', 'polite']:
        if cond not in subset.index:
            subset.loc[cond] = [np.nan] * len(valid_dims)
    subset = subset.reindex(['blunt', 'polite'])
    subset.columns = dim_labels
 
    x = np.arange(len(dim_labels))
    ax.bar(x - width/2, subset.loc['blunt'],  width, label='Blunt',  color='#e07b54', alpha=0.85)
    ax.bar(x + width/2, subset.loc['polite'], width, label='Polite', color='#5b8db8', alpha=0.85)
 
    ax.set_title(domain.capitalize(), fontsize=11)
    ax.set_xticks(x)
    ax.set_xticklabels(dim_labels, fontsize=8)
    ax.set_ylim(3.5, 5.2)
    ax.axhline(y=5, color='gray', linestyle='--', alpha=0.4)
    if ax == axes[0]:
        ax.set_ylabel('Mean Score (1–5)')
 
axes[-1].legend(title='Condition', loc='lower right')
plt.tight_layout()
plt.savefig('data/scores_by_domain.png', dpi=150)
plt.show()
```

```python
# ── Cell 3: T-tests per domain ────────────────────────────────────────────────
 
print("Paired t-tests by domain (polite vs blunt):\n")
for domain in sorted(domains):
    print(f"── {domain.upper()} ──")
    sub = df[df['domain'] == domain]
    agg = sub.groupby(['prompt_id', 'condition'])[valid_dims].mean().reset_index()
    pol = agg[agg['condition'] == 'polite'].set_index('prompt_id')
    blt = agg[agg['condition'] == 'blunt'].set_index('prompt_id')
 
    # Only run t-test if we have matched pairs
    shared = pol.index.intersection(blt.index)
    if len(shared) < 2:
        print("  Not enough prompts for t-test\n")
        continue
 
    for dim, label in zip(valid_dims, dim_labels):
        p_scores = pol.loc[shared, dim].dropna()
        b_scores = blt.loc[shared, dim].dropna()
        if len(p_scores) < 2 or p_scores.std() == 0:
            print(f"  {label:15s}  insufficient variance")
            continue
        t, p = stats.ttest_rel(p_scores, b_scores)
        diff = p_scores - b_scores
        d = diff.mean() / diff.std() if diff.std() > 0 else float('nan')
        sig = "  *" if p < 0.05 else ""
        print(f"  {label:15s}  t={t:.3f}  p={p:.3f}  Cohen's d={d:.3f}{sig}")
    print()
```

```python
# Just show descriptive means by domain and condition — no inference
domain_means = df.groupby(['domain', 'condition'])[valid_dims].mean().round(2)
print(domain_means)
```

```python
# ── Setup: Install and load fonts ─────────────────────────────────────────────

import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
from scipy import stats
import requests, os

# Download Playfair Display and Lora from Google Fonts
fonts = {
    "PlayfairDisplay-Bold.ttf":   "https://github.com/google/fonts/raw/main/ofl/playfairdisplay/PlayfairDisplay%5Bwght%5D.ttf",
    "Lora-Regular.ttf":           "https://github.com/google/fonts/raw/main/ofl/lora/Lora%5Bwght%5D.ttf",
}

font_dir = os.path.expanduser("~/.matplotlib/fonts")
os.makedirs(font_dir, exist_ok=True)

for fname, url in fonts.items():
    fpath = os.path.join(font_dir, fname)
    if not os.path.exists(fpath):
        print(f"Downloading {fname}...")
        r = requests.get(url)
        with open(fpath, "wb") as f:
            f.write(r.content)

# Rebuild font cache and register fonts
fm.fontManager.__init__()
for fname in fonts:
    fm.fontManager.addfont(os.path.join(font_dir, fname))

print("Fonts loaded ✓")


# ── Substack theme ────────────────────────────────────────────────────────────

BLUE_LIGHT  = "#73BFF9"
BLUE_DARK   = "#0284C7"
BG          = "#FAFAFA"
GRAY        = "#A9A9A9"
TEXT        = "#1a1a2e"

def apply_theme(ax, fig):
    fig.patch.set_facecolor(BG)
    ax.set_facecolor(BG)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color(GRAY)
    ax.spines['bottom'].set_color(GRAY)
    ax.tick_params(colors=TEXT, labelsize=10)
    ax.yaxis.label.set_color(TEXT)
    ax.xaxis.label.set_color(TEXT)
    for label in ax.get_xticklabels() + ax.get_yticklabels():
        label.set_fontfamily("Lora")
        label.set_color(TEXT)


# ── Cell 1: Overall bar chart ─────────────────────────────────────────────────

valid_dims = ['scores_accuracy', 'scores_completeness', 'scores_depth']
dim_labels = ['Accuracy', 'Completeness', 'Depth']
width = 0.35

means = df.groupby('condition')[valid_dims].mean()
means.index = ['Blunt', 'Polite']
means.columns = dim_labels

fig, ax = plt.subplots(figsize=(9, 5.5))
apply_theme(ax, fig)

x = np.arange(len(dim_labels))
bars1 = ax.bar(x - width/2, means.loc['Blunt'],  width, label='Blunt',  color=GRAY,       alpha=0.9, zorder=3)
bars2 = ax.bar(x + width/2, means.loc['Polite'], width, label='Polite', color=BLUE_DARK,  alpha=0.9, zorder=3)

ax.set_title('Does Politeness Change Response Quality?',
             fontfamily='Playfair Display', fontsize=15, color=TEXT, pad=14)
ax.set_ylabel('Mean Score (1–5)', fontfamily='Lora', fontsize=11)
ax.set_xticks(x)
ax.set_xticklabels(dim_labels, fontfamily='Lora', fontsize=11)
ax.set_ylim(3.5, 5.3)
ax.axhline(y=5, color=GRAY, linestyle='--', alpha=0.5, linewidth=0.8, label='Max score')
ax.yaxis.grid(True, color=GRAY, alpha=0.3, zorder=0)

for bar in bars1:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.03,
            f'{bar.get_height():.2f}', ha='center', va='bottom',
            fontfamily='Lora', fontsize=9, color=TEXT)
for bar in bars2:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.03,
            f'{bar.get_height():.2f}', ha='center', va='bottom',
            fontfamily='Lora', fontsize=9, color=TEXT)

legend = ax.legend(title='Condition', frameon=False, fontsize=10)
legend.get_title().set_fontfamily('Lora')
for text in legend.get_texts():
    text.set_fontfamily('Lora')

fig.text(0.5, -0.02, 'Clarity excluded — ceiling effect (all scores = 5)',
         ha='center', fontfamily='Lora', fontsize=8.5, color=GRAY)

plt.tight_layout()
plt.subplots_adjust(right=0.88)
plt.savefig('data/overall_scores.png', dpi=180, bbox_inches='tight')
plt.show()


# ── Cell 2: Split by domain ───────────────────────────────────────────────────

domains = sorted(df['domain'].unique())
fig, axes = plt.subplots(1, len(domains), figsize=(15, 5.5), sharey=True)
fig.patch.set_facecolor(BG)
fig.suptitle('Response Quality by Domain and Condition',
             fontfamily='Playfair Display', fontsize=15, color=TEXT, y=1.02)

for ax, domain in zip(axes, domains):
    subset = df[df['domain'] == domain].groupby('condition')[valid_dims].mean()
    for cond in ['blunt', 'polite']:
        if cond not in subset.index:
            subset.loc[cond] = [np.nan] * len(valid_dims)
    subset = subset.reindex(['blunt', 'polite'])
    subset.columns = dim_labels

    apply_theme(ax, fig)
    x = np.arange(len(dim_labels))
    ax.bar(x - width/2, subset.loc['blunt'],  width, label='Blunt',  color=GRAY,      alpha=0.9, zorder=3)
    ax.bar(x + width/2, subset.loc['polite'], width, label='Polite', color=BLUE_DARK, alpha=0.9, zorder=3)

    ax.set_title(domain.capitalize(),
                 fontfamily='Playfair Display', fontsize=12, color=TEXT)
    ax.set_xticks(x)
    ax.set_xticklabels(dim_labels, fontfamily='Lora', fontsize=9)
    ax.set_ylim(3.5, 5.3)
    ax.axhline(y=5, color=GRAY, linestyle='--', alpha=0.4, linewidth=0.8)
    ax.yaxis.grid(True, color=GRAY, alpha=0.3, zorder=0)
    if ax == axes[0]:
        ax.set_ylabel('Mean Score (1–5)', fontfamily='Lora', fontsize=11)

legend = axes[-1].legend(title='Condition', frameon=False, fontsize=9,
                          loc='upper right')
legend.get_title().set_fontfamily('Lora')
for text in legend.get_texts():
    text.set_fontfamily('Lora')

plt.tight_layout()
plt.savefig('data/scores_by_domain.png', dpi=180, bbox_inches='tight')
plt.show()
```
